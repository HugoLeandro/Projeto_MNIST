# Sobre o Projeto MNIST

## Introdução
Este projeto explora o clássico conjunto de dados MNIST, um marco na história do aprendizado de máquina, para criar modelos que reconhecem dígitos manuscritos. Ele foi desenvolvido como parte do meu portfólio pessoal para demonstrar habilidades em redes neurais, ajuste de hiperparâmetros e visualização de desempenho.


## Objetivo
O objetivo principal deste projeto foi criar e comparar diferentes arquiteturas de redes neurais para analisar como cada abordagem impacta a acurácia e eficiência do modelo. Além disso, utilizei o Keras Tuner para otimizar hiperparâmetros, explorando combinações que maximizassem os resultados.

## Estrutura do Projeto
1. **Primeiro Modelo (Rede Neural Densa)**
   - Desenvolvi um modelo inicial simples para criar uma linha de base.
   - Este modelo serviu como introdução ao uso do TensorFlow para classificação de imagens.

2. **Modelos CNN (Convolutional Neural Networks)**
   - Três arquiteturas diferentes foram implementadas para investigar como a complexidade e os ajustes impactam a capacidade do modelo de aprender representações de imagem.

3. **Otimização com Keras Tuner**
   - Utilizei o Keras Tuner para experimentar com hiperparâmetros, como número de camadas, tamanhos de filtros e taxa de dropout.
   - Este passo foi essencial para compreender o equilíbrio entre performance e overfitting.

## Ferramentas e Tecnologias
- **TensorFlow/Keras:** Para construção, treinamento e avaliação dos modelos.
- **Keras Tuner:** Para ajustar os hiperparâmetros.
- **Matplotlib/Seaborn:** Para criar gráficos que destacam a evolução do treinamento e os resultados obtidos.

## Resultados Obtidos
- **Acurácia:** O modelo com melhor desempenho alcançou uma acurácia de 100% no conjunto de teste.
- **Insights:** Redes convolucionais provaram ser mais eficazes do que redes densas para dados visuais, especialmente devido à capacidade de extrair características espaciais.

## Reflexões
Este projeto foi uma experiência valiosa que não só consolidou meu entendimento de redes neurais, mas também me desafiou a pensar criticamente sobre escolhas de design e técnicas de otimização. Ele reflete meu compromisso com a busca de soluções eficazes e bem fundamentadas em ciência de dados.
